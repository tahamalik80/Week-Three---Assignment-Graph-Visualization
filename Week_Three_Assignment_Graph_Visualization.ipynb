```python
# Importing necessary libraries
import networkx as nx   # For creating and analyzing the graph
import matplotlib.pyplot as plt  # For visualizing the graph

# Load the Karate Club graph
# NetworkX has the Karate Club graph as a built-in dataset
G = nx.karate_club_graph()

# Basic Information about the Graph
print("Number of Nodes:", G.number_of_nodes())  # Total nodes in the graph
print("Number of Edges:", G.number_of_edges())  # Total edges connecting the nodes

# Calculate Graph Diameter
# Diameter is the longest shortest path between any two nodes
diameter = nx.diameter(G)
print("Graph Diameter:", diameter)

# Calculate Average Clustering Coefficient
# This measures the degree to which nodes tend to cluster together
avg_clustering = nx.average_clustering(G)
print("Average Clustering Coefficient:", avg_clustering)

# Calculate Degree Distribution
# Degree distribution shows the number of connections each node has
degree_distribution = [d for n, d in G.degree()]
print("Degree Distribution:", degree_distribution)

# Visualize the Graph using Matplotlib
# Setting the layout for the graph
pos = nx.spring_layout(G, seed=42)  # Spring layout for better visualization

# Drawing the graph with customizations
nx.draw(G, pos, with_labels=True, node_color='skyblue',
        node_size=700, font_size=10, edge_color='gray')

# Adding a title to the graph
plt.title("Karate Club Graph Visualization")
plt.show()

# Export the Graph to GEXF format for Gephi
# GEXF format is supported by Gephi for advanced visualization
nx.write_gexf(G, "karate_club_graph.gexf")

# End of Code
```
